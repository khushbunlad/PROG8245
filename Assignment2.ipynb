##### Statistical Approaches for Outlier Detection : Z-Score Method

```python
import numpy as np
from scipy import stats

# Calculate Z-scores
z_scores = stats.zscore(Boston_dataset)
abs_z_scores = np.abs(z_scores)
threshold = 3 
Outliers_ZScore = (abs_z_scores > threshold)

# Identify outliers
Outliers_ZScore_indices = np.where(Outliers_ZScore)
print("Z-Score Outliers Indices:", Outliers_ZScore_indices)
```

##### Statistical Approaches for Outlier Detection : IQR Method

```python
Q1 = Boston_dataset.quantile(0.25)
Q3 = Boston_dataset.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = ((Boston_dataset < (Q1 - 1.5 * IQR)) | (Boston_dataset > (Q3 + 1.5 * IQR)))

outliers_iqr_indices = outliers_iqr.any(axis=1)
print("IQR Outliers Indices:", Boston_dataset[outliers_iqr_indices])
```

##### Multivariate Outlier Detection

```python
import numpy as np
from scipy.spatial import distance

# Calculate Mahalanobis distance

# Covariance matrix of the dataset
cov_matrix = np.cov(Boston_dataset.T)
# Inverse covariance matrix
inv_cov_matrix = np.linalg.inv(cov_matrix) 
# Mean of the dataset
mean = Boston_dataset.mean(axis=0)  
m_dists = np.array([distance.mahalanobis(x, mean, inv_cov_matrix) for x in Boston_dataset.values])

# Set a threshold for detection
threshold_mahalanobis = 3  # This can be adjusted based on your specific use case

# Identify multivariate outliers
outliers_mahalanobis = np.where(m_dists > threshold_mahalanobis)[0]
print("Mahalanobis Outliers Indices:", outliers_mahalanobis)
```

#### Question 7: Implementing the Local Outlier Factor (LOF) Function (25)


- Local outlier factor is a function that labels outlier based on 2 predefined parameters, the first one is t **k** for k-Nearest-Neighbours. The second one is the threshold of outliers which could be set arbitrary. However, in this coding task you are required to BOX-PLOTS to determine the threshold for the local outlier factor and label your outliers.
- You are expected to implement different functions to reach the target which will be defined below.


```python
import math


def calculate_point_to_point_distance(point_a, point_b, distance_approach="Euclidean"):
    ## TODO: This function takes as an input two points (where each point is a pair of (x,y)) and returns the distance between the points.
    ## The distance is calculate based on given distance_approach
    ## You need to handle the Euclidean and Manhattan Distance Approaches in your code
    x1, y1 = point_a
    x2, y2 = point_b
    if distance_approach == "Euclidean":
        return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    elif distance_approach == "Manhattan":
        return abs(x2 - x1) + abs(y2 - y1)
    else:
        return 0
# Helper function
def get_k_nearest_neighbors(point, list_of_points, K, distance_approach):
    distances = [(other_point, calculate_point_to_point_distance(point, other_point, distance_approach)) 
                 for other_point in list_of_points if other_point != point]
    distances.sort(key=lambda x: x[1])
    kth_distance = distances[K-1][1]
    
    # Neighborhood set of points within Nd distance
    neighborhood = [pt for pt, dist in distances if dist <= kth_distance]
    return neighborhood, kth_distance

# Helper function
def reachability_distance(point, neighbor, kth_distance):
    # Reachability distance calculation as per provided steps
    return max(kth_distance, calculate_point_to_point_distance(point, neighbor))

# Helper function
def local_reachability_density(point, list_of_points, K, distance_approach):
    neighborhood, kth_distance = get_k_nearest_neighbors(point, list_of_points, K, distance_approach)
    reach_distances = [reachability_distance(point, neighbor, kth_distance) for neighbor in neighborhood]
    return len(neighborhood) / sum(reach_distances) if reach_distances else float('inf')

# Helper function
def calculate_lof(point, list_of_points, K, distance_approach):
    neighborhood, _ = get_k_nearest_neighbors(point, list_of_points, K, distance_approach)
    point_lrd = local_reachability_density(point, list_of_points, K, distance_approach)
    
    # Sum of LRD values for neighbors and reachability distances
    neighbor_lrd_sum = sum(local_reachability_density(neighbor, list_of_points, K, distance_approach) for neighbor in neighborhood)
    reach_distances_sum = sum(reachability_distance(point, neighbor, calculate_point_to_point_distance(point, neighbor, distance_approach)) 
                              for neighbor in neighborhood)
    
    return (neighbor_lrd_sum * reach_distances_sum) / (len(neighborhood) ** 2 * point_lrd) if point_lrd > 0 else float('inf')

# Helper function
def get_points_lof(list_of_points, K=2, distance_approach = "Euclidean"):
    ## TODO: This function takes a list of points and returns a dictionary
    ## The dictionary is in format {point: LOF, ...}
    ## You can use as much helper functions as you'd like
    list_of_points = [tuple(point) for point in list_of_points]  # Ensure all points are tuples
    lof_scores = {point: calculate_lof(point, list_of_points, K, distance_approach) for point in list_of_points}
    return lof_scores

def get_outliers(list_of_points, K=2, distance_approach="Euclidean"):
    ## TODO: This function uses the get_points_lof to get the LOF scores, and uses is to build a box-plot
    ## Using the box-plot points are labelled as outliers are returned to the user
    ## Example: if point (2,1) and (3,2) are labelled as outliers using box-plot
    ## Then you will return the example given in the place-holder below

    # LOF scores for each point
    lof_scores = get_points_lof(list_of_points, K, distance_approach)

    # Calculate IQR
    lof_values = np.array(list(lof_scores.values()))
    q1 = np.percentile(lof_values, 25)
    q3 = np.percentile(lof_values, 75)
    iqr = q3 - q1
    upper_whisker = q3 + 1.5 * iqr

    outliers = [point for point, lof in lof_scores.items() if lof > upper_whisker]
    return outliers 
```

##### To calculate the LOF you need to apply the following steps:
 1. Calculate the distance between each pair of points p1,p2
 2. Find your Kth nearest neighbor point (x) from current point (p):
    - After finding it, remember the distance *Nd* which is the distance from p to x, and save a set of elements which are in a neighborhood of *Nd* around point p.
    - Save the length of this set of elements.
 3. Calculate ReachDistance(x,p)=max(Nd_x, Distance(x,p))
 3. Calculate LRD (Local Reachability Distance) where LRD_p = length_of_set_p/Sum[ReachDistance(x,p)] (for each x in set of elements of point p)
 4. Calculate LOF (Local Outlier Factor) where LOF_p = {Sum[LRD_x] (for each x in set of elements of point p) * Sum[RD(x,p)] (for each x in set of elements of point p)} / length_of_set_p^2
